#1.In languages, the order of the words and their position in a sentence really matters. The meaning of the entire sentence can change if the words are re-ordered.

#2.When implementing NLP solutions, recurrent neural networks have an inbuilt mechanism that deals with the order of sequences. The transformer model, however, does not use recurrence or convolution and treats each data point as independent of the other.

#3.Hence, positional information is added to the model explicitly to retain the information regarding the order of words in a sentence!

##**What Is Positional Encoding?**

##1.Positional encoding describes the location or position of an entity in a sequence so that each position is assigned a unique representation.

##2.There are many reasons why a single number, such as the index value, is not used to represent an item’s position in transformer models. For long sequences, the indices can grow large in magnitude. If you normalize the index value to lie between 0 and 1, it can create problems for variable length sequences as they would be normalized differently.

##3.Transformers use a smart positional encoding scheme, where each position/index is mapped to a vector. Hence, the output of the positional encoding layer is a matrix, where each row of the matrix represents an encoded object of the sequence summed with its positional information.
